```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data for algorithms
algorithms = ["FIFO", "Optimal", "LRU", "LFU", "MFU"]

# Ten frames test cases
ten_frames_data = {
    "FIFO": [12, 8, 13, 10, 8],
    "Optimal": [11, 8, 11, 10, 8],
    "LRU": [13, 8, 11, 10, 8],
    "LFU": [13, 8, 11, 12, 8],
    "MFU": [11, 8, 12, 12, 8]
}

# Twenty frames test cases
twenty_frames_data = {
    "FIFO": [23, 22, 30, 30, 23],
    "Optimal": [21, 21, 23, 22, 22],
    "LRU": [21, 22, 27, 23, 24],
    "LFU": [21, 22, 27, 23, 24],
    "MFU": [23, 23, 32, 26, 24]
}

# Frame sizes
frame_sizes = [10, 20]

# Average and standard deviation for 10 and 20 frames
average_faults = {
    "FIFO": [np.mean(ten_frames_data["FIFO"]), np.mean(twenty_frames_data["FIFO"])],
    "Optimal": [np.mean(ten_frames_data["Optimal"]), np.mean(twenty_frames_data["Optimal"])],
    "LRU": [np.mean(ten_frames_data["LRU"]), np.mean(twenty_frames_data["LRU"])],
    "LFU": [np.mean(ten_frames_data["LFU"]), np.mean(twenty_frames_data["LFU"])],
    "MFU": [np.mean(ten_frames_data["MFU"]), np.mean(twenty_frames_data["MFU"])]
}
std_dev = {
    "FIFO": [np.std(ten_frames_data["FIFO"]), np.std(twenty_frames_data["FIFO"])],
    "Optimal": [np.std(ten_frames_data["Optimal"]), np.std(twenty_frames_data["Optimal"])],
    "LRU": [np.std(ten_frames_data["LRU"]), np.std(twenty_frames_data["LRU"])],
    "LFU": [np.std(ten_frames_data["LFU"]), np.std(twenty_frames_data["LFU"])],
    "MFU": [np.std(ten_frames_data["MFU"]), np.std(twenty_frames_data["MFU"])]
}

# 1. Bar Chart: Page Faults vs Algorithms for 10 and 20 Frames
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Bar chart for 10 frames
x = np.arange(len(algorithms))
ten_frames_means = [average_faults[algo][0] for algo in algorithms]
axes[0].bar(x, ten_frames_means, color='skyblue')
axes[0].set_title("Page Faults vs Algorithms (10 Frames)")
axes[0].set_xticks(x)
axes[0].set_xticklabels(algorithms)
axes[0].set_ylabel("Page Faults")

# Bar chart for 20 frames
twenty_frames_means = [average_faults[algo][1] for algo in algorithms]
axes[1].bar(x, twenty_frames_means, color='salmon')
axes[1].set_title("Page Faults vs Algorithms (20 Frames)")
axes[1].set_xticks(x)
axes[1].set_xticklabels(algorithms)

plt.tight_layout()
plt.show()

# 2. Line Charts: Page Faults vs Frame Size for Various Algorithms
# Combined line chart
plt.figure(figsize=(8, 6))
for algo in algorithms:
    plt.plot(frame_sizes, average_faults[algo], marker='o', label=algo)
    for i, frame in enumerate(frame_sizes):
        plt.text(frame, average_faults[algo][i], f"{frame}", fontsize=9, ha='center', va='bottom')
plt.title("Page Faults vs Frame Size (All Algorithms)")
plt.xlabel("Frame Size")
plt.ylabel("Average Page Faults")
plt.xticks(frame_sizes)
plt.legend()
plt.grid()
plt.show()

# Individual line charts for each algorithm
fig, axes = plt.subplots(3, 2, figsize=(12, 12), sharex=True, sharey=True)
axes = axes.ravel()

for i, algo in enumerate(algorithms):
    axes[i].plot(frame_sizes, average_faults[algo], marker='o', label=algo, color='tab:blue')
    for j, frame in enumerate(frame_sizes):
        axes[i].text(frame, average_faults[algo][j], f"{frame}", fontsize=9, ha='center', va='bottom')
    axes[i].set_title(f"Page Faults vs Frame Size ({algo})")
    axes[i].set_xlabel("Frame Size")
    axes[i].set_ylabel("Average Page Faults")
    axes[i].grid()
    axes[i].legend()

# Hide any unused subplot
for j in range(len(algorithms), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()

# 3. Generate Table: Average and Standard Deviation
data = {
    "Algorithm": algorithms,
    "Avg Faults (10 Frames)": [average_faults[algo][0] for algo in algorithms],
    "Std Dev (10 Frames)": [std_dev[algo][0] for algo in algorithms],
    "Avg Faults (20 Frames)": [average_faults[algo][1] for algo in algorithms],
    "Std Dev (20 Frames)": [std_dev[algo][1] for algo in algorithms]
}
df = pd.DataFrame(data)

# Print Table
print("Table: Average and Standard Deviation for Page Faults")
print(df)

# Save table to a CSV file
df.to_csv("average_std_table.csv", index=False)
print("\nTable saved as 'average_std_table.csv'.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data for algorithms
algorithms = ["FIFO", "Optimal", "LRU", "LFU", "MFU"]

# Average and standard deviation for 10 and 20 frames
average_faults = {
    "FIFO": [10.2, 8.4],
    "Optimal": [9.6, 8.4],
    "LRU": [10.0, 8.4],
    "LFU": [10.4, 8.4],
    "MFU": [10.2, 8.4]
}
std_dev = {
    "FIFO": [2.04, 1.67],
    "Optimal": [1.48, 1.67],
    "LRU": [1.79, 1.67],
    "LFU": [2.17, 1.67],
    "MFU": [1.92, 1.67]
}

# 4. Importance of Standard Deviation
# Create a grouped bar chart to show average vs standard deviation
x = np.arange(len(algorithms))  # X positions for the groups
width = 0.35  # Width of the bars

# Data for 10 frames
avg_10_frames = [average_faults[algo][0] for algo in algorithms]
std_10_frames = [std_dev[algo][0] for algo in algorithms]

# Plotting the bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bars_avg = ax.bar(x - width / 2, avg_10_frames, width, label="Average Page Faults", color='skyblue')
bars_std = ax.bar(x + width / 2, std_10_frames, width, label="Standard Deviation", color='orange')

# Add labels and title
ax.set_title("Average Page Faults vs Standard Deviation (10 Frames)", fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(algorithms)
ax.set_ylabel("Page Faults")
ax.legend()

# Adding values on top of bars
for bar in bars_avg:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, f"{yval:.2f}", ha="center", fontsize=9)
for bar in bars_std:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, f"{yval:.2f}", ha="center", fontsize=9)

plt.tight_layout()
plt.show()
```
